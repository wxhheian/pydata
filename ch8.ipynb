### Database-Style DataFrame Joins

```python
#数据集的合并（merge）或连接（join）运算是通过一个或多个键将行链接起来的
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                    'data2': range(3)})
```

```python
df1
```

```python
df2
```

```python
#多对一的合并，df1中的数据有多个被标记为a和b行，而df2中key列的每个值则仅对应一行
pd.merge(df1,df2)
```

```python
pd.merge(df2,df1)  #结果与pd.merge(df1,df2)一样
#注意对于两个dataframe中key列没有重复的值，就删除
```

```python
#注意上面没有指明用哪个列进行连接，如果没有指定，merge会将重叠的列名当作键
#也可以指定
pd.merge(df1,df2,on='key')
```

```python
#如果两个dataframe的列名不同，也可以分别指定
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                    'data2': range(3)})
```

```python
df3
```

```python
df4
```

```python
pd.merge(df3,df4,left_on='lkey',right_on='rkey')  #df3为left,df4为right
```

```python
#注意里面c和d以及与之相关的数据都消失了
#默认情况下，merge做的是内连接，结果中的键是交集
#merge还有外连接，结果是键的并集，组合了左连接和右连接
pd.merge(df1,df2,how='outer')  
#how = inner,left,right  #使用两个表都有的键，左表的键盘，右表的键
```

```python
pd.merge(df1,df2,how='left')
```

```python
#多对多的合并有些不直观
#每个dataframe都有多个a,b,c行
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                    'data1': range(6)})
df2 = pd.DataFrame({'key': ['a', 'b', 'a', 'b', 'd'],
                    'data2': range(5)})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='key',how='left')
```

```python
pd.merge(df1,df2,how='inner')
```

```python
#多对多的连接产生的是行的笛卡尔积，左边的dataframe有3个b行，右边dataframe
#有2个b行，所有最终结果就有6个b行
```

### 重塑层次化索引

```python
#stack:将数据的列旋转为行
#unstack:将数据的行旋转为列
data = pd.DataFrame(np.arange(6).reshape((2, 3)),
                    index=pd.Index(['Ohio', 'Colorado'], name='state'),
                    columns=pd.Index(['one', 'two', 'three'],
                    name='number'))
#pd.Index
```

```python
data
```

```python
result = data.stack()   #结果返回层次化索引的series
#列变成了内层的行索引
```

```python
result
```

```python
result.unstack()  #将内层的行索引变成列
```

```python
#默认下，stack和unstack操作的都是最内层
#传入分层级别的编号或名称可以对不同级别进行操作
result.unstack(0)
```

```python
result.unstack('state')
```

### Pivoting “Long” to “Wide” Format

```python
#多个时间序列通常是以长格式（long）或堆叠格式（stacked）存储在数据库或csv中
data = pd.read_csv('examples/macrodata.csv')
```

```python
data.head()
```

```python
#pd.PeriodIndex
periods = pd.PeriodIndex(year=data.year,quarter=data.quarter,name='date')
```

```python
periods
```

```python
columns = pd.Index(['realgdp', 'infl', 'unemp'], name='item')
```

```python
columns
```

```python
data = data.reindex(columns=columns)
```

```python
data.index = periods.to_timestamp('D', 'end')
#pandas.Period.to_timestamp(freq,how)
```

```python
data.head()
```

```python
data.stack()
```

```python
data.stack().reset_index()
```

```python
ldata = data.stack().reset_index().rename(columns={0:'values'})
```

```python
#上面的结果就是关系型数据库经常存储的方式
```

```python
#将不同item值分别形成一列，date列中的时间戳作为索引
#pivot 在轴上转动
pivoted = ldata.pivot('date','item','values')
#第1个参数作为行索引，第2个参数作为列索引，最后一个可选参数作为填充dataframe的数据列
```

```python
pivoted
```

```python
#为ldata增加一列
ldata['value2'] = np.random.randn(len(ldata))
```

```python
ldata[:10]
```

```python
#如果忽略最后一个参数，得到的dataframe就会带有层次化的列
pivoted = ldata.pivot('date','item')
```

```python
pivoted
```

```python
pivoted['values'][:5]
```

```python
#pivot其实是用set_index创建层次化索引，再用unstack重塑
unstacked = ldata.set_index(['date','item']).unstack('item')
unstacked
```

### Pivoting “Wide” to “Long” Format

```python
melted = pd.melt(df,['key'])   #转换为long format
```

```python
melted
```

```python
reshaped = melted.pivot('key','variable','value')
```

```python
reshaped
```

```python
reshaped.reset_index()  
#pivot的结果从列创建了一个索引，用作行标签，可以使用reset_index将数据移回列
```
