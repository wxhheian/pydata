```python
#csv文件的形式很多，只需要定义csv.Dialect的一个子类即可定义出新的格式
class my_dialect(csv.Dialect):
    lineterminator = '\n'
    delimiter = ';'
    quotechar = '"'
    quoting = csv.QUOTE_MINIMAL
#reader = csv.reader(f,dialect=my_dialect)   #f为open函数返回的文件对象
```

```python
#定义csv文件也可以传入参数给csv.reader
#reader = csv.reader(f,delimiter='|')
```

```python
#csv.writer与csv.reader相反
with open('mydata.csv','w') as f:
    writer = csv.writer(f,dialect=my_dialect)
    writer.writerow(('one','two','three'))
    writer.writerow(('1','2','3'))
    writer.writerow(('4','5','6'))
    writer.writerow(('7','8','9'))
    
```

```python
from lxml import objectify

path = 'datasets/mta_perf/Performance_MNR.xml'
parsed = objectify.parse(open(path))
root = parsed.getroot()
```

```python
root
```

```python
data = []

skip_fields = ['PARENT_SEQ', 'INDICATOR_SEQ',
               'DESIRED_CHANGE', 'DECIMAL_PLACES']

for elt in root.INDICATOR:
    el_data = {}
    for child in elt.getchildren():
        if child.tag in skip_fields:
            continue
        el_data[child.tag] = child.pyval
    data.append(el_data)
```

```python
data
```

```python
perf = pd.DataFrame(data)
```

### 使用HDF5

```python
#HDF5是一种存储大规模科学数组数据的非常好的文件格式
#HDF5(hierarchical data format)
```

```python
frame = pd.DataFrame({'a':np.random.randn(100)})
frame
```

```python
store = pd.HDFStore('mydata.h5')
```

```python
store['obj1'] = frame
store['obj1_col'] = frame['a']  #HDF5可以存储多个数据集
store
```

```python
#HDF5文件中的对象可以像字典一样获取
store['obj1']
```

```python
#HDFStore支持两种存储模式:fixed和table,后者更慢，但支持使用特殊语法进行查询操作
store.put('obj2',frame,format='table')
```

```python
store['obj2']
```

```python
store.select('obj2',where=['index >= 10 and index <= 15'])
```

```python
#put是store['obj2'] =frame方法显示版本，它允许我们设置其他选项，比如格式
#HDF5可以存储多个数据集
```

```python
store['obj1_col']  #series对象
```

```python
store.close()
```

```python
#pandas.read_hdf可以读取hdf5文件
#pandas.to_hdf 将pandas数据写入hdf5文件
frame.to_hdf('mydata.h5','obj3',format='table')
```

## Interacting with Databases

```python
import sqlite3  #python内置的sqlite3驱动器
```

```python
query = """
CREATE TABLE test
(a VARCHAR(20), b VARCHAR(20),
 c REAL,        d INTEGER
);"""
con = sqlite3.connect('mydata.sqlite')  #位置在/pydata
pydata
con.execute(query)
con.commit()
```

```python
data = [('Atlanta', 'Georgia', 1.25, 6),
        ('Tallahassee', 'Florida', 2.6, 3),
        ('Sacramento', 'California', 1.7, 5)]
stmt = "INSERT INTO test VALUES(?, ?, ?, ?)"
con.executemany(stmt, data)
con.commit()
```

```python
#从数据库中选取数据，大部分python SQL驱动器都会返回一个元组列表
cursor = con.execute('select * from test')
```

```python
rows = cursor.fetchall()
```

```python
rows
```

```python
cursor.description   #列名在光标的description属性中
```

```python
pd.DataFrame(rows,columns=[x[0] for x in cursor.description])
```
